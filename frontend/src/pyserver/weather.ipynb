```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
casual_df=pd.read_csv('Casualties0514.csv')
accident_df=pd.read_csv('Accidents0514.csv')
vehicle_df=pd.read_csv('Vehicles0514.csv')
```

```python
first_df=pd.merge(casual_df,accident_df,on='Accident_Index')
```

```python
df=pd.merge(first_df,vehicle_df,on='Accident_Index')
```

```python
df.drop('LSOA_of_Accident_Location',axis=1,inplace=True)
```

```python
df.dropna(subset=['Time'],axis=0,inplace=True)
df
```

```python
df.isnull().values.any()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().values.any()
df.columns
```

```python
q3_df=pd.DataFrame(data=df,columns=['Longitude','Latitude','1st_Road_Class','2nd_Road_Class','Weather_Conditions'])
```

```python
q3_df
```

```python
X=q3_df.drop('Weather_Conditions',axis=1)
y=q3_df['Weather_Conditions']

q3_df['Weather_Conditions'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y)
X_train.head(20)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree= DecisionTreeClassifier()
```

```python
dtree.fit(X_train.values,y_train.values)
```

```python
predictions= dtree.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test.values,predictions))
```

```python
print(confusion_matrix(y_test.values,predictions))
q3_df
```
